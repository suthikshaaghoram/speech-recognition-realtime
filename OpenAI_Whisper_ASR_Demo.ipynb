```python

# load audio and pad/trim it to fit 30 seconds
audio = whisper.load_audio("/content/santhanam.mp3")
audio = whisper.pad_or_trim(audio)

# make log-Mel spectrogram and move to the same device as the model
mel = whisper.log_mel_spectrogram(audio).to(model.device)

# detect the spoken language
_, probs = model.detect_language(mel)
print(f"Detected language: {max(probs, key=probs.get)}")

# decode the audio
options = whisper.DecodingOptions()
result = whisper.decode(model, mel, options)

# print the recognized text
print(result.text)
```

```python
import gradio as gr
import time
```

```python
def transcribe(audio):
    # Load and preprocess audio
    audio = whisper.load_audio(audio)
    audio = whisper.pad_or_trim(audio)

    # Convert to log-Mel spectrogram
    mel = whisper.log_mel_spectrogram(audio).to(model.device)

    # Detect the spoken language
    _, probs = model.detect_language(mel)
    detected_lang = max(probs, key=probs.get)
    print(f"Detected language: {detected_lang}")

    # Decode to get original transcription
    decode_options = whisper.DecodingOptions()
    result_original = whisper.decode(model, mel, decode_options)

    # Decode with translation task to get English translation
    translate_options = whisper.DecodingOptions(task="translate")
    result_translated = whisper.decode(model, mel, translate_options)

    return f"Detected Language: {detected_lang}\nOriginal Text: {result_original.text}\nEnglish Translation: {result_translated.text}"


```

```python

gr.Interface(
    title = 'OpenAI Whisper ASR Gradio Web UI',
    fn=transcribe,
    inputs=[
        gr.Audio(sources="microphone", type="filepath")
    ],
    outputs=[
        "textbox"
    ],
    live=True).launch()
```
